```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
rc('font', family='Malgun Gothic')  # 윈도우 환경의 경우
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 깨짐 방지

# 데이터 정의
categories = [
    "급여 (62.7%)",
    "성장 가능성 (39.1%)",
    "워라밸 (39.1%)",
    "동료와의 팀워크 (17.3%)",
    "상급자 및 동료로부터의 인정 (11.5%)"
]
scores = [62.7, 39.1, 39.1, 17.3, 11.5]

# 데이터 정렬 (높은 값부터 낮은 값으로)
sorted_data = sorted(zip(scores, categories), key=lambda x: x[0], reverse=True)
scores, categories = zip(*sorted_data)

# 색상 정의
colors = ['#0033cc', '#0055ff', '#0077ff', '#3399ff', '#66b2ff']

# 그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.barh(categories, scores, color=colors[:len(categories)], edgecolor='black')

# 그래프 제목 및 레이블 설정
plt.title("직장 생활 만족을 위한 '가치 요인 조사'", fontsize=16, pad=20)
plt.xlabel("응답 비율 (%)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# 점수 표시
for bar, score in zip(bars, scores):
    plt.text(score + 1, bar.get_y() + bar.get_height() / 2,
             f'{score}%', va='center', fontsize=10, color='black')

# 그래프 표시
plt.gca().invert_yaxis()  # Y축 뒤집기 (높은 점수가 위로 오도록)
plt.tight_layout()
plt.show()
```
