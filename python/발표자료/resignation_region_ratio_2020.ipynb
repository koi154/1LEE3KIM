```python
data_2020 = data2020.iloc[:, 0:5]
data_2020 = data_2020.drop(index=[0, 1])
# data_2014 = data_2014.set_index("지역별(1)")
data_2020 = data_2020.rename(
    columns={"2022": "합계", "2022.1": "경력자", "2022.2": "입사 경력자", "2022.3": "신입"})
part = ["경력", "입사\n경험", "신입"]
data_2020

# data_2022.to_excel("./output/지역별 퇴사 인원.xlsx")
```

```python
seoul = data_2020.iloc[1]
values_list = []
total = 0

for i in range(2,len(seoul)):
    value = seoul.iloc[i]
    values_list.append(value)
    total += int(value)
```

```python
seoul = data_2020.iloc[1]
values_list = []
total = 0
for i in range(2,len(seoul)):
    value = seoul.iloc[i]
    values_list.append(value)
    total += int(value)

seoul_ratio = []
for i in range(len(values_list)):
    seoul_ratio.append(values_list[i]/total*100)
seoul_ratio

font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc("font", family=font_name)

wedgeprops = {"width": 0.7, "edgecolor": "k", "linewidth": 1}
plt.pie(seoul_ratio, labels=part, autopct="%.1f%%", colors=[
        "#B26600", "#4169E1", "#FFD700"], wedgeprops=wedgeprops, explode=[0.05, 0.05, 0.05], textprops={"color": "black", "fontsize": 14, "fontweight": "bold", "ha": "center"}, shadow=True, startangle=90, pctdistance=0.7,
        labeldistance=1.2)
plt.title("2020 서울특별시 퇴사율")
plt.show()
```
