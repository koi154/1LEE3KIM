```python
data_2014 = data2014.iloc[:, 0:5]
data_2014 = data_2014.drop(index=[0, 1])
# data_2014 = data_2014.set_index("지역별(1)")
data_2014 = data_2014.rename(
    columns={"2014.1": "경력자", "2014.2": "입사 경력자", "2014.3": "신입"})

data_2014
```

```python
data_2014.iloc[1]
list = []
for i in range(1, len(data_2014.iloc[1])):
    list.append(data_2014.iloc[1][i])
seoul_ratio = []

for i in range(1,len(list)):
    seoul_ratio.append(list[i]/list[0]*100)
seoul_ratio
```

```python
seoul_part = ["경력", "입사\n경험", "신입"]
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc("font", family=font_name)

wedgeprops = {"width": 0.7, "edgecolor": "k", "linewidth": 1}
plt.pie(seoul_ratio, labels=seoul_part, autopct="%.1f%%", colors=[
        "#B26600", "#4169E1", "#FFD700"], wedgeprops=wedgeprops, explode=[0.05, 0.05, 0.05], textprops={"color": "black", "fontsize": 14, "fontweight": "bold", "ha": "center"}, shadow=True, startangle=90, pctdistance=0.7,
        labeldistance=1.2)
plt.show()
```
