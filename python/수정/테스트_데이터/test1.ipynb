```python
# resignation = data[["2012.1","2012.2"]]
list = [data[["지역별(1)", "2014.1", "2014.2","2014.3"]]]
for i in range(2015, 2023):
    list.append(data[[f"{i}.1", f"{i}.2", f"{i}.3"]])
df_list = [pd.DataFrame(item) for item in list]

result_df = pd.concat(df_list, axis=1)
result_df = result_df.drop(0, axis=0)
result_df = result_df.rename(columns=lambda x: x.replace("지역별(1)", "지역"))
result_df.to_excel("./output/지역별_퇴사자3.xlsx", index=False)
# result_df.to_csv("./output/지역별_퇴사자2.cvs", index=False)
```

```python
data3 = pd.read_excel("./output/지역별_퇴사자2.xlsx", index_col=0)
data3
plt.figure(figsize=(25, 15))
sns.set(font="Malgun Gothic", rc={"axes.unicode_minus": False})
data_row = data3.iloc[1]
# data_row
colors = ["#B26600", "#4169E1", "#FFD700"]
x_labels = []
for i in range(2014, 2023):
    x_labels.append(f"{i}_경력")
    x_labels.append(f"입사경험")
    x_labels.append(f"신입")
sns.barplot(x=data_row.index, y=data_row.values, palette=colors)
plt.xlabel("연도")
plt.ylabel("퇴사자 수")
plt.xticks(range(len(x_labels)), x_labels, rotation=0)
plt.title("연도별 퇴사자 수 (2014~2022)")
```
