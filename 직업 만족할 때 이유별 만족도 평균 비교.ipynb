```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
rc('font', family='Malgun Gothic')  # 윈도우의 경우 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 깨짐 방지

# 데이터 정의
categories = [
    "연봉이 높은 직업군이라서",
    "일이 재미있고 흥미로워서",
    "타고난 성격성향과 잘 맞아서",
    "사회적으로 인정받을 수 있어서"
]
scores = [8.3, 7.4, 7, 6]

# 색상 정의 (항목별 다른 색상 지정)
colors = ['#FFD700', '#FFC107', '#FF9800', '#FF5722']

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.barh(categories, scores, color=colors)

# 그래프 제목 및 축 레이블 설정
plt.title('직업 만족할 때 이유별 만족도 평균 비교', fontsize=16)
plt.xlabel('평균 점수', fontsize=12)
plt.ylabel('이유', fontsize=12)

# 점수 표시
for index, value in enumerate(scores):
    plt.text(value + 0.1, index, f'{value}점', va='center', fontsize=10)

# 그래프 표시
plt.gca().invert_yaxis()  # Y축 뒤집기 (상위 카테고리가 위로 오도록)
plt.tight_layout()
plt.show()
```
