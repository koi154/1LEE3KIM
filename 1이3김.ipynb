```python
import matplotlib.pyplot as plt
from matplotlib import rc

#    (Windows: Malgun Gothic, Mac: AppleGothic, Ubuntu: NanumGothic)
rc('font', family='Malgun Gothic')  # Windows 
# rc('font', family='AppleGothic')  # Mac 
# rc('font', family='NanumGothic')  # Ubuntu 

# 
categories = ["   ", "   ", "    "]
values = [35.3, 3, 59.5]
explode = (0, 0, 0.05)  # '    '  

#           
combined_percentage = values[0] + values[1]

#   
plt.figure(figsize=(8, 8))
plt.pie(
    values,
    labels=categories,
    autopct='%1.1f%%',  #  
    startangle=90,      #  
    colors=plt.cm.Paired.colors,  #  
    shadow=True,       #  
    explode=explode,    # explode  
    textprops={'fontsize': 17}
)

#      
plt.gca().set_aspect('equal')  #   
plt.figtext(0.5, -0.1, f"    : {combined_percentage:.1f}%", 
            ha='center', fontsize=15, color='black', weight='bold')

plt.title('2020     ', fontsize=18)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

#      
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows   Malgun Gothic 
# MacOS 'AppleGothic'    

#  
data = {
    " ": [
        "   ", " ()", "", 
        " ", "   ", " ,   ", 
        " "
    ],
    "": [45.9, 36.2, 31.5, 21.4, 20.3, 18.4, 17.4]
}

# pandas DataFrame 
df = pd.DataFrame(data)

#    
df_sorted = df.sort_values(by='', ascending=False)

#    (  )
plt.figure(figsize=(11, 6))
bars = plt.bar(df_sorted[' '], df_sorted[''], color='skyblue')

#    
plt.ylabel(' (%)',fontsize=13)
plt.title('2022   ', fontsize=20)
plt.xticks(rotation=30, ha='right',fontsize = 13)  # x   
plt.yticks(fontsize=13)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

#     
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  #   x 
        bar.get_height() - 5,              #    (   )
        f'{bar.get_height()}%',            #  
        ha='center', va='center', color='blue', fontsize=15  #    
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#  
years = ['2016', '2018', '2020', '2021']
score = [2.93, 3.01, 2.9, 3.1]  #   (1~5)
satisfaction_rate = [22.8, 27.4, 20, 20.4]  #   (%)
additional_data = [27.8, 28.5, 33.9, 33.9]  #   (   )

#  
fig, ax1 = plt.subplots()

# x  
x_pos = np.arange(len(years))

#     ( )
bar1 = ax1.bar(x_pos - 0.2, score, color='orange', label=' ', width=0.4)

#   y  
ax1.set_ylabel('')  # y  
ax1.yaxis.set_ticks([])  # y  

#   y  2.4 3.3  ( )
ax1.set_ylim(2.4, 3.3)

#   y      
ax2 = ax1.twinx()

#     ( )
bar2 = ax2.bar(x_pos + 0.2, satisfaction_rate, color='blue', alpha=0.5, label='  ', width=0.4)

#   y   (   y)
ax2.set_ylabel('   (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.set_ylim(15, 30)

#   y    
ax3 = ax1.twinx()  #  y 
ax3.spines['right'].set_position(('outward', 60))  #  120    

#   (     )
line, = ax3.plot(x_pos, additional_data, color='green', marker='o', linestyle='-', label='  ', linewidth=2)

#   y   (  y)
ax3.set_ylabel('  ', color='green')
ax3.tick_params(axis='y', labelcolor='green')
ax3.set_ylim(20, 40)

# x   
ax1.set_xticks(x_pos)
ax1.set_xticklabels(years)

#  
plt.title('  ,   ')

#   (  )
fig.tight_layout()  #  
#   label     
ax1.legend(handles=[bar1], loc='upper left', bbox_to_anchor=(0.0, 1))
ax2.legend(handles=[bar2], loc='upper left', bbox_to_anchor=(0.0, 0.9))
ax3.legend(handles=[line], loc='upper right', bbox_to_anchor=(1, 1))  #     

plt.show()
```
