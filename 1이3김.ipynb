```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정 (Windows: Malgun Gothic, Mac: AppleGothic, Ubuntu: NanumGothic)
rc('font', family='Malgun Gothic')  # Windows의 경우
# rc('font', family='AppleGothic')  # Mac의 경우
# rc('font', family='NanumGothic')  # Ubuntu의 경우

# 데이터
categories = ["스스로 그만둔 적이 있다", "해고를 당한 적이 있다", "둘 다 경험한 적이 없다"]
values = [35.3, 3, 59.5]
explode = (0, 0, 0.05)  # '둘 다 경험한 적이 없다' 항목만 강조

# 스스로 그만두거나 해고당한 비율과 둘 다 경험한 적 없다 비율 합산
combined_percentage = values[0] + values[1]

# 파이 차트 생성
plt.figure(figsize=(8, 8))
plt.pie(
    values,
    labels=categories,
    autopct='%1.1f%%',  # 퍼센트 표시
    startangle=90,      # 시작 각도
    colors=plt.cm.Paired.colors,  # 색상 팔레트
    shadow=True,       # 그림자 설정
    explode=explode,    # explode 인자 추가
    textprops={'fontsize': 17}
)

# 합산된 비율을 차트 아래에 텍스트로 추가
plt.gca().set_aspect('equal')  # 차트를 원형으로 유지
plt.figtext(0.5, -0.1, f"스스로 그만 두거나 해고당한 비율: {combined_percentage:.1f}%", 
            ha='center', fontsize=15, color='black', weight='bold')

plt.title('2020 퇴사 경험 여부 설문 결과', fontsize=18)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 예시 데이터
years = ['2016', '2018', '2020', '2021']
score = [2.93, 3.01, 2.9, 3.1]  # 점수 평가 (1~5점)
satisfaction_rate = [22.8, 27.4, 20, 20.4]  # 만족 비율 (%)
additional_data = [27.8, 28.5, 33.9, 33.9]  # 퇴사이유 값 (선형 그래프에 사용될 데이터)

# 그래프 그리기
fig, ax1 = plt.subplots()

# x축 위치 설정
x_pos = np.arange(len(years))

# 첫 번째 막대 그래프 (점수 평가)
bar1 = ax1.bar(x_pos - 0.2, score, color='orange', label='평균 점수', width=0.4)

# 첫 번째 y축 레이블을 제거
ax1.set_ylabel('')  # y축 레이블 제거
ax1.yaxis.set_ticks([])  # y축 눈금 제거

# 첫 번째 y축의 범위를 2.4에서 3.3으로 제한 (점수 평가용)
ax1.set_ylim(2.4, 3.3)

# 두 번째 y축을 추가하여 만족 비율과 선형 그래프 그리기
ax2 = ax1.twinx()

# 두 번째 막대 그래프 (만족 비율)
bar2 = ax2.bar(x_pos + 0.2, satisfaction_rate, color='blue', alpha=0.5, label='임금 만족 비율', width=0.4)

# 두 번째 y축 범위 설정 (만족 비율을 위한 y축)
ax2.set_ylabel('임금 만족 비율 (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.set_ylim(15, 30)

# 세 번째 y축을 추가하여 선형 그래프 그리기
ax3 = ax1.twinx()  # 새로운 y축 생성
ax3.spines['right'].set_position(('outward', 60))  # 오른쪽으로 120만큼 이동시켜서 겹치지 않게 설정

# 선형 그래프 (추가적인 다른 값에 대한 선형 그래프)
line, = ax3.plot(x_pos, additional_data, color='green', marker='o', linestyle='-', label='임금 불만 퇴사율', linewidth=2)

# 세 번째 y축 범위 설정 (선형 그래프용 y축)
ax3.set_ylabel('임금 불만 퇴사율', color='green')
ax3.tick_params(axis='y', labelcolor='green')
ax3.set_ylim(20, 40)

# x축의 카테고리 이름 설정
ax1.set_xticks(x_pos)
ax1.set_xticklabels(years)

# 그래프 제목
plt.title('임금 만족 비율, 임금 불만 퇴사율')

# 범례 표시 (그래프 안에 넣기)
fig.tight_layout()  # 레이아웃 조정
# 각 축의 label을 명시적으로 설정하여 모든 범례를 포함
ax1.legend(handles=[bar1], loc='upper left', bbox_to_anchor=(0.0, 1))
ax2.legend(handles=[bar2], loc='upper left', bbox_to_anchor=(0.0, 0.9))
ax3.legend(handles=[line], loc='upper right', bbox_to_anchor=(1, 1))  # 우측 상단으로 퇴사율 범례 이동

plt.show()
```
